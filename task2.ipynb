```python
def large_num(a,b,c):
    if(a>=b and a>=c):
        return a
    elif(b>a and b>c):
        return b
    else:
        return c
a = int(input("Enter num1: "))
b = int(input("Enter num2: "))
c = int(input("Enter num3: "))
print("largest number: ",large_num(a,b,c))
```

```python
def leng_func(str):
    cnt = 0
    for s in str:
        cnt +=1
    return cnt
str = input("Enter string: ")
print("length is: ",leng_func(str))
```

```python
def func(n):
    s = 1
    cnt = 0
    for s in range(1,n):
        if n%s == 0:
            cnt +=1
    if cnt == 1:
        return 1
    else:
        return 0
        
n = int(input("Enter number: "))
val = func(n)
if val == 1:
    print("prime")
else:
    print("Not prime")
```

```python
def fib(n):
    if n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fib(n-1)+fib(n-2)

n = int(input("Enter num: "))
print("Nth fib number: ",fib(n))
```

```python
def rev_str(s):
    if len(s) == 0:
        return s
    else:
        return rev_str(s[1:]) + s[0]

s = input("Enter string: ")
print("Reversed string: ",rev_str(s))
```
