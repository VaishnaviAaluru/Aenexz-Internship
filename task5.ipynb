```python
import pandas as pd
import seaborn as sns
```

```python
titanic = pd.read_csv("Titanic-Dataset.csv")
```

```python
titanic.head()
```

```python
filtered_data = titanic[(titanic["Age"] >= 18) & (titanic["Fare"] > 30)]
```

```python
filtered_data.head()
```

```python
titanic["family_size"] = titanic["SibSp"] + titanic["Parch"] + 1
```

```python
titanic["fareP"] = titanic["Fare"] / titanic["family_size"]
```

```python
titanic[["family_size", "fareP"]].head()
```

```python
avg_fare = titanic.groupby(["Pclass", "Sex"])["Fare"].mean()
avg_fare
```

```python
survival_rate = titanic.groupby(["Pclass", "Sex"])["Survived"].mean()
survival_rate
```
