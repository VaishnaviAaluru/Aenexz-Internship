```python
import random
sec_num = random.randint(1,100)
print("The Guess Game")
while True:
    g = int(input("Enter number: "))
    if g < sec_num:
        print("Your number less than the secret number\n")
    elif g > sec_num:
        print("Your number greater than the secret number\n")
    else:
        print("This is the right numberr")
```
